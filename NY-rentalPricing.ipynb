```python
for col in df.columns:
  unique_vals = df[col].unique()
  print(f'Unique values in column {col}:{unique_vals}')
```

```python
for column in df.columns:
  unique_values = df[column] .nunique()
  value_counts = df[column] .value_counts()
  print(f'Column name: {column}')
  print(f'Number of unique values: {unique_values}')
  print(f'Value counts: \n{value_counts}\n')
```

```python
#Grafico do preÃ§o
plt.figure(figsize=(10, 8))
sns.histplot(x= 'price', data=df, bins=50)
plt.title('Grafico do preÃ§o')
plt.show()
```

```python
#Grafico dos tipos de quartos
plt.figure(figsize=(10 ,8))
sns.countplot(x ='room_type', data = df)
plt.title('Room types')
plt.show()
```

```python
#Grafico de dispersÃ£o de preÃ§o x noites minimas
plt.figure(figsize=(10, 8))
sns.scatterplot(x='minimum_nights', y='price', data=df)
plt.title('PreÃ§o x Noites minimas')
plt.show()
```

```python
#matriz de correlÃ§ao
coor = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(coor, annot=True, cmap='coolwarm')
plt.title('Matriz de correlaÃ§Ã£o')
plt.show()
```
